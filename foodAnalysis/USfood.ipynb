```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'historicalcpi.csv'
historical_cpi_df = pd.read_csv(file_path)
```

```python
sns.set_style("whitegrid")
```

```python
average_cpi_change_by_year = historical_cpi_df.groupby('Year')['Percent change'].mean().reset_index()
```

```python
# Visualization 1: Line plot of the average percent change in CPI for all items over the years
average_cpi_change_by_year = historical_cpi_df.groupby('Year')['Percent change'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=average_cpi_change_by_year, x='Year', y='Percent change', marker='o')
plt.title('Average Annual Percent Change in CPI Over Time')
plt.xlabel('Year')
plt.ylabel('Average Percent Change in CPI')
plt.show()
```

```python

# Visualization 2: Bar plot of the average percent change in CPI for the most recent year available in the dataset
most_recent_year = historical_cpi_df['Year'].max()
cpi_change_most_recent_year = historical_cpi_df[historical_cpi_df['Year'] == most_recent_year]

plt.figure(figsize=(12, 6))
sns.barplot(data=cpi_change_most_recent_year, x='Consumer Price Index item', y='Percent change', color='skyblue')
plt.title(f'Percent Change in CPI by Item for {most_recent_year}')
plt.xlabel('Consumer Price Index Item')
plt.ylabel('Percent Change in CPI')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=historical_cpi_df, x='Year', y='Percent change')
plt.title('Distribution of Annual Percent Changes in CPI Over Time')
plt.xlabel('Year')
plt.ylabel('Percent Change in CPI')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization 4: Trend line for each CPI item over time
plt.figure(figsize=(14, 7))
for item in historical_cpi_df['Consumer Price Index item'].unique():
    # Filter the DataFrame for each item and plot
    item_df = historical_cpi_df[historical_cpi_df['Consumer Price Index item'] == item]
    if item_df.shape[0] > 1:  # Only plot if there is more than one data point
        sns.lineplot(data=item_df, x='Year', y='Percent change', label=item)

plt.title('Trend of Percent Change in CPI for Each Item Over Time')
plt.xlabel('Year')
plt.ylabel('Percent Change in CPI')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Move the legend out of the plot
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to fit legend
plt.show()
```

```python
# Visualization 6: Area plot for top 5 items with the most change over time
top_items = historical_cpi_df.groupby('Consumer Price Index item')['Percent change'].std().sort_values(ascending=False).head(5).index
plt.figure(figsize=(14, 7))
for item in top_items:
    item_df = historical_cpi_df[historical_cpi_df['Consumer Price Index item'] == item]
    plt.fill_between(item_df['Year'], item_df['Percent change'], label=item, alpha=0.5)

plt.title('Top 5 Items with Most CPI Change Over Time')
plt.xlabel('Year')
plt.ylabel('Percent Change in CPI')
plt.legend()
plt.show()
```

```python
top_items_data = historical_cpi_df[historical_cpi_df['Consumer Price Index item'].isin(top_items)]
top_items_pivot = top_items_data.pivot(index='Year', columns='Consumer Price Index item', values='Percent change').fillna(0)
top_items_pivot.plot(kind='bar', stacked=True, figsize=(14, 7))
plt.title('Stacked Bar Plot of Percent Change for Top 5 CPI Items by Year')
plt.xlabel('Year')
plt.ylabel('Percent Change in CPI')
plt.legend(title='Consumer Price Index item')
plt.show()
```
